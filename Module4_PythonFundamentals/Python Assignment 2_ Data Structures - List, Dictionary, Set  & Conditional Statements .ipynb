Sets (Operations):

```python
#a. Create a set called my_set with following values:['a','e','i','o','u','a','a','i'] Analyse the output and provide explanation
my_set = {'a','e','i','o','u','a','a','i'}
my_set
```

When we create a set with {'a','e','i','o','u','a','a','i'}, the output only shows {'a','e','i','o','u'} because:

-Sets do not allow duplicate elements, repeated values like 'a' and 'i' are automatically removed.

-Only unique distinct values are stored in a set.

-Sets are unordered collections the elements may not appear in the same order as they were added.

```python
#b. Attempt to change the value of my_set[4] = 's'. If code throws an error, provide an explanation.
my_set.add[4] = 's'
```

-Sets are unordered collections and do not support indexing or item assignment. Thatâ€™s why my_set[4] = 's' causes an error. To add a new element, we use the .add() method instead.

-The correct syntax for adding an element to a set is .add(value), not .add[index].

```python
#Performance Category Program:
#1. Prompt user for Input. Score range should be from 0 to 10 (both inclusive).
score = int(input('Enter your score (0 to 10):'))

if score > 7:
  print (f"Performance_category: Above Average")
  print("Excellent work! Keep up.")

elif  4 <= score <= 7:
  print(f"Performance_category: Average")
  print("Good job!")

else:
      print("Performance_category: Below Average")
      print("Need to improve your performance.")
```
